# Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Collect Data

```python
#membaca data covid
covid = pd.read_csv('data/covid_indo.csv')
#membuang kolom code, benua, dan lokasi/negara
covid = covid.drop(columns=['iso_code','continent','location'])
#mengubah tipe kolom Date menjadi tipe datetime
covid['date'] = pd.to_datetime(covid['date'])
#menampilkan 5 baris data pertama
covid.head()
```

```python
#membaca data GAS
GAS = pd.read_csv('data/GAS.csv')
#mengambil kolom Date dan Price
GAS = GAS[['Date','Price']].copy()
#mengubah Price menjadi GAS dan Date menjadi date
GAS.rename(columns={'Price':'gas/mmbtu','Date':'date'}, inplace=True)
#mengubah tipe kolom date menjadi tipe datetime
GAS['date'] = pd.to_datetime(GAS['date'])
#mengubah tipe kolom gas/mmbtu menjadi float
GAS['gas/mmbtu'] = GAS['gas/mmbtu'].astype(float)
#membulatkan harga gas/mmbtu sehingga hanya memiliki 2 angka setelah koma
GAS['gas/mmbtu'] = GAS['gas/mmbtu'].round(2)
#menampilkan 5 baris data pertama
GAS.head()
```

```python
#membaca data MINYAK
MINYAK = pd.read_csv('data/MINYAK.csv')
#mengambil kolom Date dan Price
MINYAK = MINYAK[['Date','Price']].copy()
#mengubah Price menjadi MINYAK dan Date menjadi date
MINYAK.rename(columns={'Price':'oil/bbl','Date':'date'}, inplace=True)
#mengubah tipe kolom date menjadi tipe datetime
MINYAK['date'] = pd.to_datetime(MINYAK['date'])
#mengubah tipe kolom MINYAK/bbl menjadi float
MINYAK['oil/bbl'] = MINYAK['oil/bbl'].astype(float)
#membulatkan harga MINYAK/bbl sehingga hanya memiliki 2 angka setelah koma
MINYAK['oil/bbl'] = MINYAK['oil/bbl'].round(2)
#menampilkan 5 baris data pertama
MINYAK.head()
```

```python
#membaca data PERAK
PERAK = pd.read_csv('data/PERAK.csv')
#mengambil kolom Date dan Price
PERAK = PERAK[['Date','Price']].copy()
#mengubah Price menjadi PERAK dan Date menjadi date
PERAK.rename(columns={'Price':'silver/gr','Date':'date'}, inplace=True)
#mengubah tipe kolom date menjadi tipe datetime
PERAK['date'] = pd.to_datetime(PERAK['date'])
#mengubah tipe kolom PERAK/gr menjadi float
PERAK['silver/gr'] = PERAK['silver/gr'].astype(float)
#mengubah harga emas/gr yang awalnya harga/1 troy ons menjadi harga/1 gram
PERAK['silver/gr'] = PERAK['silver/gr'] /31.1034768
#membulatkan harga PERAK/gr sehingga hanya memiliki 2 angka setelah koma
PERAK['silver/gr'] = PERAK['silver/gr'].round(2)
#menampilkan 5 baris data pertama
PERAK.tail()
```

```python
#membaca data EMAS
EMAS = pd.read_csv('data/EMAS.csv')
#mengambil kolom Date dan Price
EMAS = EMAS[['Date','Price']].copy()
#mengubah Price menjadi EMAS dan Date menjadi date
EMAS.rename(columns={'Price':'gold/gr','Date':'date'}, inplace=True)
#mengubah tipe kolom date menjadi tipe datetime
EMAS['date'] = pd.to_datetime(EMAS['date'])
#menghilangkan koma dalam kolom gold/gr
EMAS['gold/gr'] = EMAS['gold/gr'].str.replace(',', '')
#mengubah tipe kolom gold/gr menjadi float
EMAS['gold/gr'] = EMAS['gold/gr'].astype(float)
#mengubah harga gold/gr yang awalnya harga/1 troy ons menjadi harga/1 gram
EMAS['gold/gr'] = EMAS['gold/gr'] * 100/31.1034768
#membulatkan harga gold/gr sehingga hanya memiliki 2 angka setelah koma
EMAS['gold/gr'] = EMAS['gold/gr'].round(2)
#menampilkan 5 baris data pertama
EMAS.head()
```

```python
#membaca data USD
USD = pd.read_csv('data/USD.csv')
#mengambil kolom Date dan Price
USD = USD[['Date','Price']].copy()
#mengubah Price menjadi USD dan Date menjadi date
USD.rename(columns={'Price':'usd','Date':'date'}, inplace=True)
#mengubah tipe kolom date menjadi tipe datetime
USD['date'] = pd.to_datetime(USD['date'])
#menghilangkan koma dalam kolom usd
USD['usd'] = USD['usd'].str.replace(',', '')
#mengubah tipe kolom usd menjadi float
USD['usd'] = USD['usd'].astype(float)
#menampilkan 5 baris data pertama
USD.head()
```

```python
#membaca data EUR
EUR = pd.read_csv('data/EUR.csv')
#mengambil kolom Date dan Price
EUR = EUR[['Date','Price']].copy()
#mengubah Price menjadi EUR dan Date menjadi date
EUR.rename(columns={'Price':'eur','Date':'date'}, inplace=True)
#mengubah tipe kolom date menjadi tipe datetime
EUR['date'] = pd.to_datetime(EUR['date'])
#menghilangkan koma dalam kolom EUR
EUR['eur'] = EUR['eur'].str.replace(',', '')
#mengubah tipe kolom EUR menjadi float
EUR['eur'] = EUR['eur'].astype(float)
#menampilkan 5 baris data pertama
EUR.head()
```

```python
def convert_to_million(value):
    if value.endswith('B'):
        return float(value.replace('B', '')) * 1000
    elif value.endswith('M'):
        return float(value.replace('M', ''))
```

```python
#membaca data IHSG
IHSG = pd.read_csv('data/IHSG.csv')
#mengambil kolom Date dan Price
IHSG = IHSG[['Date','Price','Vol.']].copy()
#mengubah Price menjadi ihsg dan Date menjadi date
IHSG.rename(columns={'Price':'ihsg','Date':'date','Vol.':'vol ihsg'}, inplace=True)
#mengubah tipe kolom date menjadi tipe datetime
IHSG['date'] = pd.to_datetime(IHSG['date'])
#menghilangkan koma dalam kolom ihsg
IHSG['ihsg'] = IHSG['ihsg'].str.replace(',', '')
#mengubah tipe kolom ihsd menjadi float
IHSG['ihsg'] = IHSG['ihsg'].astype(float)
IHSG['vol ihsg'] = IHSG['vol ihsg'].astype(str)
IHSG['vol ihsg'] = IHSG['vol ihsg'].apply(convert_to_million)
#menampilkan 5 baris data pertama
IHSG.head()
```

```python
#membaca data saham
JII = pd.read_csv('data/JII.csv')
#mengambil kolom Date dan Price
JII = JII[['Date','Price','Vol.']].copy()
JII.rename(columns={'Date': 'date','Price':'jii','Vol.':'vol jii'}, inplace=True)
#mengubah tipe kolom Date menjadi tipe datetime
JII['date'] = pd.to_datetime(JII['date'])
#menampilkan 5 baris data pertama#
JII['vol jii'] = JII['vol jii'].astype(str)
JII['vol jii'] = JII['vol jii'].apply(convert_to_million)
JII.tail(8)
```

# Data Gather

```python
data = GAS.merge(covid, on='date', how='inner')
data = MINYAK.merge(data, on='date', how='inner')
data = PERAK.merge(data, on='date', how='inner')
data = EMAS.merge(data, on='date', how='inner')
data = EUR.merge(data, on='date', how='inner')
data = USD.merge(data, on='date', how='inner')
data = IHSG.merge(data, on='date', how='inner')
data = JII.merge(data, on='date', how='inner')
#membalikkan data terakhir menjadi index pertama dan mereset indexnya
data = data.iloc[::-1].reset_index(drop=True)
#mencari kolom yang isinya NaN semua
columns_with_all_nan = data.columns[data.isna().all()]
#membuang kolom yang isinya NaN semua
data = data.drop(columns=columns_with_all_nan)
#menghitung jumlah nilai unik untuk mengidentifikasi kolom-kolom dengan semua nilai yang sama
all_values_same = data.nunique() == 1
#memfilter kolom-kolom dengan semua nilai yang sama
columns_with_same_values = all_values_same[all_values_same].index.tolist()
#menghapus kolom dengan nilai yang sama
data = data.drop(columns=columns_with_same_values)
#mengisi nilai NaN dengan 0.0
data = data.fillna(0.0)
data.set_index('date', inplace=True)
data['silver/gr'] = data['silver/gr'] * data['usd']
data['oil/bbl'] = data['oil/bbl'] * data['usd']
data['gas/mmbtu'] = data['gas/mmbtu'] * data['usd']
#menampilkan 5 baris data pertama
data.head()
```

# Data Pre-Processing

```python
min_val = data.min()
max_val = data.max()
normalized_data = -1 + 2 * (data - min_val) / (max_val - min_val)
normalized_data.head()
```

```python
normalized_data.describe()
```

# Data Understanding

```python
#Mencari distribusi data
df_hist =normalized_data.copy()
df_hist.hist(figsize=(30,30))
plt.show()
```

```python
correlation = normalized_data.corrwith(normalized_data['jii'])

# Menanggulangi underscore di label
correlation.index = correlation.index.str.replace('_', ' ')

plt.figure(figsize=(20, 16), dpi=300)  # Mengatur ukuran dan DPI

# Menggunakan gaya 'ggplot'
plt.style.use('ggplot')

# Membuat plot bar
ax = correlation.plot(kind='bar', color='skyblue')

# Menambahkan label dan memutar label sumbu x
plt.xlabel('Feature', fontsize=14)
plt.ylabel('Correlation', fontsize=14)
plt.xticks(rotation=45, ha="right", fontsize=14)

# Menambahkan nilai korelasi pada setiap bar
for i, v in enumerate(correlation):
    ax.text(i, v + 0.01, str(round(v, 2)), ha='center', va='bottom', fontsize=12)

# Menyimpan gambar ke file 'korelasi.jpg'
plt.savefig('korelasi.png')

# Menampilkan plot
plt.show()
```
